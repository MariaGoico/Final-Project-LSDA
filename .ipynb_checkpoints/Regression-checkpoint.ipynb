```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, substring, when, regexp_replace, trim
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
import glob
import os

spark = SparkSession.builder \
    .appName("Analisis_Nieve_2024_Robust") \
    .master("local[*]") \
    .config("spark.sql.ansi.enabled", "false") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .config("spark.memory.offHeap.enabled", "true") \
    .config("spark.memory.offHeap.size", "2g") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")
```

```python
base_path = "/home/alumno/reposirotio/Grupo3"

# Encontrar directorio 2024
posibles_dirs = glob.glob(f"{base_path}/*2024*")
target_dir = None
for d in posibles_dirs:
    if "2024" in os.path.basename(d):
        target_dir = d
        break
if not target_dir: target_dir = f"{base_path}/2024.tar"

print(f"Procesando: {target_dir}")

try:
    # 2. CARGA Y LIMPIEZA
    print("Cargando CSVs...")
    df = spark.read.option("header", "true").option("inferSchema", "false").csv(target_dir)

    # Filtrar encabezados repetidos
    df = df.filter(col("LATITUDE") != "LATITUDE")

    target_col = "SNDP"
    numeric_cols = [
        "LATITUDE", "LONGITUDE", "ELEVATION",
        "TEMP", "TEMP_ATTRIBUTES",
        "DEWP", "DEWP_ATTRIBUTES",
        "SLP", "SLP_ATTRIBUTES",
        "STP", "STP_ATTRIBUTES",
        "VISIB", "VISIB_ATTRIBUTES",
        "WDSP", "WDSP_ATTRIBUTES",
        "MXSPD", "GUST", 
        "MAX", "MAX_ATTRIBUTES", 
        "MIN", "MIN_ATTRIBUTES",
        "PRCP"
    ]

    # Limpieza de asteriscos y conversión
    select_exprs = []
    for c in numeric_cols + [target_col]:
        # Borrar asteriscos (*) y espacios
        clean_col = regexp_replace(trim(col(c)), "[*]", "")
        
        select_exprs.append(
            when(clean_col == "", 0.0)
            .when(clean_col.cast("double") > 9000, 0.0)  # Filtra 9999.9
            .when(clean_col.cast("double") == 999.9, 0.0) # Filtra 999.9
            .otherwise(clean_col.cast("double"))
            .alias(c)
        )

    # Desglose FRSHTT
    frshtt_col = when(col("FRSHTT").isNull() | (col("FRSHTT") == ""), "000000").otherwise(col("FRSHTT"))
    frshtt_exprs = [
        substring(frshtt_col, 1, 1).cast("int").alias("Is_Fog"),
        substring(frshtt_col, 2, 1).cast("int").alias("Is_Rain"),
        substring(frshtt_col, 3, 1).cast("int").alias("Is_Snow"),
        substring(frshtt_col, 4, 1).cast("int").alias("Is_Hail"),
        substring(frshtt_col, 5, 1).cast("int").alias("Is_Thunder"),
        substring(frshtt_col, 6, 1).cast("int").alias("Is_Tornado")
    ]

    data = df.select(*select_exprs, *frshtt_exprs).na.fill(0.0)

    # 3. VECTORIZACIÓN
    input_cols = [c for c in data.columns if c != target_col and c != "features"]
    assembler = VectorAssembler(inputCols=input_cols, outputCol="features")
    final_data = assembler.transform(data)

    # 4. SPLIT Y CACHE
    train, test = final_data.randomSplit([0.8, 0.2], seed=42)
    
    # Cacheamos en memoria RAM para acelerar las 1000 iteraciones
    train.cache()
    count_train = train.count()
    print(f"Datos cacheados en RAM. Filas de entrenamiento: {count_train}")

    # 5. ENTRENAMIENTO
    if count_train > 0:
        print("Entrenando modelo optimizado (ElasticNet + 1000 iter)...")
        lr = LinearRegression(
            featuresCol="features", 
            labelCol=target_col, 
            maxIter=1000,        
            regParam=0.1,        
            elasticNetParam=0.8  
        )
        
        model = lr.fit(train)
        predictions = model.transform(test)

        # 6. RESULTADOS
        r2 = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="r2").evaluate(predictions)
        rmse = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="rmse").evaluate(predictions)

        print("\n" + "="*60)
        print(f"RESULTADOS FINALES (RMSE: {rmse:.4f} | R2: {r2:.4f})")
        print("="*60)
        
        print("VARIABLES REALMENTE IMPORTANTES (Top 10):")
        coeffs = sorted(zip(input_cols, model.coefficients), key=lambda x: abs(x[1]), reverse=True)
        
        for name, val in coeffs[:10]:
            if abs(val) > 0.00001:
                impacto = "AUMENTA NIEVE" if val > 0 else "DISMINUYE NIEVE"
                print(f"{name:.<25} {val:+.5f}  ({impacto})")
    else:
        print("Error: No quedaron datos tras el filtrado.")

except Exception as e:
    print(f"ERROR DURANTE EL PROCESO: {e}")
    import traceback
    traceback.print_exc()

finally:
    # 7. CIERRE SEGURO (Esto evita tu error Py4JError)
    try:
        # Solo intentamos parar si la sesión no está ya detenida
        if 'spark' in locals() and not spark.sparkContext._jsc.sc().isStopped():
            print("\nCerrando sesión de Spark limpiamente...")
            spark.stop()
    except Exception:
        pass # Si falla al cerrar, ignoramos el error porque ya estaba cerrado
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, substring, when, regexp_replace, trim
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
import glob
import os

# 1. INICIALIZAR (Configuración Robusta)
spark = SparkSession.builder \
    .appName("Analisis_Nieve_SOLO_NIEVE") \
    .master("local[*]") \
    .config("spark.sql.ansi.enabled", "false") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

base_path = "/home/alumno/reposirotio/Grupo3"
# Buscar 2024
posibles_dirs = glob.glob(f"{base_path}/*2024*")
target_dir = posibles_dirs[0] if posibles_dirs else f"{base_path}/2024.tar"

try:
    print(f"1. Cargando datos de: {target_dir}")
    df = spark.read.option("header", "true").option("inferSchema", "false").csv(target_dir)

    # 2. LIMPIEZA Y SELECCIÓN
    df = df.filter(col("LATITUDE") != "LATITUDE") # Quitar headers repetidos

    target_col = "SNDP"
    numeric_cols = [
        "LATITUDE", "ELEVATION", "TEMP", "DEWP", "SLP", "STP", "VISIB", "WDSP", 
        "MXSPD", "GUST", "MAX", "MIN", "PRCP"
        # Quitamos los ATTRIBUTES para limpiar ruido
    ]

    select_exprs = []
    for c in numeric_cols + [target_col]:
        clean_col = regexp_replace(trim(col(c)), "[*]", "")
        select_exprs.append(
            when(clean_col == "", 0.0)
            .when(clean_col.cast("double") > 9000, 0.0)
            .when(clean_col.cast("double") == 999.9, 0.0)
            .otherwise(clean_col.cast("double"))
            .alias(c)
        )

    # FRSHTT
    frshtt_col = when(col("FRSHTT").isNull(), "000000").otherwise(col("FRSHTT"))
    frshtt_exprs = [
        substring(frshtt_col, 3, 1).cast("int").alias("Is_Snow"), # Solo nos importa nieve hoy
        substring(frshtt_col, 2, 1).cast("int").alias("Is_Rain")
    ]

    data = df.select(*select_exprs, *frshtt_exprs).na.fill(0.0)

    # -------------------------------------------------------------
    # EL PASO CLAVE: FILTRAR SOLO DÍAS CON NIEVE (> 0)
    # -------------------------------------------------------------
    print("2. Filtrando: Nos quedamos SOLO con días que tienen nieve en el suelo...")
    snow_data = data.filter(col("SNDP") > 0)
    
    count = snow_data.count()
    print(f"   Filas encontradas con nieve real: {count}")

    if count > 100:
        # 3. VECTORIZACIÓN
        assembler = VectorAssembler(
            inputCols=[c for c in snow_data.columns if c != target_col and c != "features"], 
            outputCol="features"
        )
        final_data = assembler.transform(snow_data)

        train, test = final_data.randomSplit([0.8, 0.2], seed=42)

        # 4. ENTRENAMIENTO (Regresión Lineal Pura)
        # Usamos menos regularización aquí porque los datos ya son "buenos"
        lr = LinearRegression(featuresCol="features", labelCol=target_col, maxIter=200)
        
        print("3. Entrenando modelo de regresión en subconjunto de nieve...")
        model = lr.fit(train)
        predictions = model.transform(test)

        # 5. RESULTADOS
        r2 = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="r2").evaluate(predictions)
        rmse = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="rmse").evaluate(predictions)

        print("\n" + "="*60)
        print(f"RESULTADOS EN DÍAS NEVADOS (SNDP > 0)")
        print("="*60)
        print(f"R2 (Precisión explicativa): {r2:.4f}") 
        print(f"RMSE (Error medio): {rmse:.4f} pulgadas")
        print("-" * 60)
        
        # Interpretación automática
        if r2 > 0.4:
            print("CONCLUSIÓN: ¡ÉXITO! La regresión funciona muy bien cuando hay nieve.")
            print("El R2 bajo anterior era culpa de los días con 0 nieve.")
        else:
            print("CONCLUSIÓN: Los datos son muy ruidosos incluso con nieve.")

        print("\nFactores que aumentan la profundidad (Top 5):")
        coeffs = sorted(zip(assembler.getInputCols(), model.coefficients), key=lambda x: x[1], reverse=True)
        for name, val in coeffs[:5]:
            print(f"{name}: {val:+.4f}")

    else:
        print("No hay suficientes datos con nieve > 0 en este año para entrenar.")

except Exception as e:
    print(e)
    import traceback
    traceback.print_exc()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, substring, when, regexp_replace, trim
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.evaluation import RegressionEvaluator
import glob
import os

# 1. INICIALIZAR (Configuración Robusta 4GB)
spark = SparkSession.builder \
    .appName("Analisis_Nieve_RandomForest") \
    .master("local[*]") \
    .config("spark.sql.ansi.enabled", "false") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

base_path = "/home/alumno/reposirotio/Grupo3"
# Buscar 2024
posibles_dirs = glob.glob(f"{base_path}/*2024*")
target_dir = posibles_dirs[0] if posibles_dirs else f"{base_path}/2024.tar"

print(f"Procesando con RANDOM FOREST: {target_dir}")

try:
    # 2. CARGA Y LIMPIEZA (Igual que antes)
    df = spark.read.option("header", "true").option("inferSchema", "false").csv(target_dir)
    df = df.filter(col("LATITUDE") != "LATITUDE")

    target_col = "SNDP"
    numeric_cols = [
        "LATITUDE", "ELEVATION", "TEMP", "DEWP", "SLP", "STP", "VISIB", "WDSP", 
        "MXSPD", "GUST", "MAX", "MIN", "PRCP"
    ]

    select_exprs = []
    for c in numeric_cols + [target_col]:
        clean_col = regexp_replace(trim(col(c)), "[*]", "")
        select_exprs.append(
            when(clean_col == "", 0.0)
            .when(clean_col.cast("double") > 9000, 0.0)
            .when(clean_col.cast("double") == 999.9, 0.0)
            .otherwise(clean_col.cast("double"))
            .alias(c)
        )

    frshtt_col = when(col("FRSHTT").isNull(), "000000").otherwise(col("FRSHTT"))
    frshtt_exprs = [
        substring(frshtt_col, 1, 1).cast("int").alias("Is_Fog"),
        substring(frshtt_col, 2, 1).cast("int").alias("Is_Rain"),
        substring(frshtt_col, 3, 1).cast("int").alias("Is_Snow"),
        substring(frshtt_col, 4, 1).cast("int").alias("Is_Hail"),
        substring(frshtt_col, 5, 1).cast("int").alias("Is_Thunder"),
        substring(frshtt_col, 6, 1).cast("int").alias("Is_Tornado")
    ]

    data = df.select(*select_exprs, *frshtt_exprs).na.fill(0.0)

    # 3. VECTORIZACIÓN
    input_cols = [c for c in data.columns if c != target_col and c != "features"]
    assembler = VectorAssembler(inputCols=input_cols, outputCol="features")
    final_data = assembler.transform(data)

    # 4. SPLIT
    train, test = final_data.randomSplit([0.8, 0.2], seed=42)
    
    # 5. ENTRENAMIENTO CON RANDOM FOREST
    print("Entrenando Random Forest (esto puede tardar unos minutos)...")
    
    # Configuración de Random Forest:
    # numTrees=50:  Crea 50 árboles de decisión y promedia sus resultados (más robusto).
    # maxDepth=10:  Permite que los árboles sean profundos para capturar matices complejos.
    # maxBins=32:   Necesario para discretizar variables continuas.
    rf = RandomForestRegressor(
        featuresCol="features", 
        labelCol=target_col,
        numTrees=50, 
        maxDepth=10, 
        seed=42
    )
    
    model = rf.fit(train)
    predictions = model.transform(test)

    # 6. RESULTADOS
    r2 = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="r2").evaluate(predictions)
    rmse = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="rmse").evaluate(predictions)

    print("\n" + "="*60)
    print(f"RESULTADOS RANDOM FOREST")
    print("="*60)
    print(f"R2 (Precisión): {r2:.4f}")
    print(f"RMSE (Error): {rmse:.4f}")
    print("-" * 60)
    
    print("IMPORTANCIA DE VARIABLES (Feature Importance):")
    # Random Forest tiene una propiedad 'featureImportances' muy útil
    importances = model.featureImportances
    
    # Mapear importancia a nombres de columnas
    feat_imp_list = []
    for i, col_name in enumerate(input_cols):
        feat_imp_list.append((col_name, importances[i]))
    
    # Ordenar y mostrar
    feat_imp_list.sort(key=lambda x: x[1], reverse=True)
    
    for name, imp in feat_imp_list[:10]:
        print(f"{name:.<25} {imp:.4f} ({imp*100:.1f}%)")

except Exception as e:
    print(f"ERROR: {e}")
    import traceback
    traceback.print_exc()

finally:
    if 'spark' in locals() and not spark.sparkContext._jsc.sc().isStopped():
        spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, substring, when, regexp_replace, trim, to_date, dayofyear
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import GBTRegressor  # <--- EL NUEVO MOTOR
from pyspark.ml.evaluation import RegressionEvaluator
import glob
import os

# 1. INICIALIZAR (Configuración Robusta 4GB)
spark = SparkSession.builder \
    .appName("Analisis_Nieve_GBT_Time") \
    .master("local[*]") \
    .config("spark.sql.ansi.enabled", "false") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

base_path = "/home/alumno/reposirotio/Grupo3"
# Buscar 2024
posibles_dirs = glob.glob(f"{base_path}/*2024*")
target_dir = posibles_dirs[0] if posibles_dirs else f"{base_path}/2024.tar"

print(f"Mejorando modelo con GBT y Tiempo en: {target_dir}")

try:
    # 2. CARGA
    df = spark.read.option("header", "true").option("inferSchema", "false").csv(target_dir)
    df = df.filter(col("LATITUDE") != "LATITUDE")

    # --- MEJORA 1: INGENIERÍA DE CARACTERÍSTICAS (TIEMPO) ---
    # Convertimos la fecha (MM/dd/yyyy) a un número del 1 al 365
    # Esto ayuda al modelo a entender la estacionalidad (Invierno vs Verano)
    if "DATE" in df.columns:
        df = df.withColumn("ParsedDate", to_date(col("DATE"), "MM/dd/yyyy"))
        df = df.withColumn("DayOfYear", dayofyear(col("ParsedDate")))
    else:
        # Fallback por si acaso, aunque DATE debería estar
        df = df.withColumn("DayOfYear", (col("TEMP") * 0) + 180) 

    target_col = "SNDP"
    numeric_cols = [
        "DayOfYear", # <--- NUEVA VARIABLE ESTRELLA
        "LATITUDE", "ELEVATION", "TEMP", "DEWP", "SLP", "STP", "VISIB", "WDSP", 
        "MXSPD", "GUST", "MAX", "MIN", "PRCP"
    ]

    select_exprs = []
    # Incluimos DayOfYear en la limpieza
    for c in numeric_cols + [target_col]:
        clean_col = regexp_replace(trim(col(c)), "[*]", "")
        select_exprs.append(
            when(clean_col == "", 0.0)
            .when(clean_col.cast("double") > 9000, 0.0)
            .when(clean_col.cast("double") == 999.9, 0.0)
            .otherwise(clean_col.cast("double"))
            .alias(c)
        )

    # FRSHTT
    frshtt_col = when(col("FRSHTT").isNull(), "000000").otherwise(col("FRSHTT"))
    frshtt_exprs = [
        substring(frshtt_col, 1, 1).cast("int").alias("Is_Fog"),
        substring(frshtt_col, 2, 1).cast("int").alias("Is_Rain"),
        substring(frshtt_col, 3, 1).cast("int").alias("Is_Snow"),
        substring(frshtt_col, 4, 1).cast("int").alias("Is_Hail"),
        substring(frshtt_col, 5, 1).cast("int").alias("Is_Thunder"),
        substring(frshtt_col, 6, 1).cast("int").alias("Is_Tornado")
    ]

    data = df.select(*select_exprs, *frshtt_exprs).na.fill(0.0)

    # 3. VECTORIZACIÓN
    input_cols = [c for c in data.columns if c != target_col and c != "features"]
    assembler = VectorAssembler(inputCols=input_cols, outputCol="features")
    final_data = assembler.transform(data)

    # 4. SPLIT
    train, test = final_data.randomSplit([0.8, 0.2], seed=42)
    
    # 5. ENTRENAMIENTO CON GBT (Gradient Boosted Trees)
    print("Entrenando GBT (Esto es más lento que Random Forest, paciencia)...")
    
    # maxIter=50: Hará 50 pasadas de corrección.
    # stepSize=0.1: Aprende despacio para no 'memorizar' (overfitting).
    gbt = GBTRegressor(
        featuresCol="features", 
        labelCol=target_col, 
        maxIter=50,
        maxDepth=8,
        stepSize=0.1,
        seed=42
    )
    
    model = gbt.fit(train)
    predictions = model.transform(test)

    # 6. RESULTADOS
    r2 = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="r2").evaluate(predictions)
    rmse = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="rmse").evaluate(predictions)

    print("\n" + "="*60)
    print(f"RESULTADOS GBT + FEATURE ENGINEERING")
    print("="*60)
    print(f"R2 (Precisión): {r2:.4f}")
    print(f"RMSE (Error): {rmse:.4f}")
    print("-" * 60)
    
    print("IMPORTANCIA DE VARIABLES (GBT):")
    importances = model.featureImportances
    feat_imp_list = sorted(zip(input_cols, importances), key=lambda x: x[1], reverse=True)
    
    for name, imp in feat_imp_list[:10]:
        print(f"{name:.<25} {imp:.4f} ({imp*100:.1f}%)")

except Exception as e:
    print(f"ERROR: {e}")
    import traceback
    traceback.print_exc()

finally:
    if 'spark' in locals() and not spark.sparkContext._jsc.sc().isStopped():
        spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, substring, when, regexp_replace, trim, to_date, dayofyear
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import DecisionTreeRegressor  # <--- EL VELOCISTA
from pyspark.ml.evaluation import RegressionEvaluator
import glob
import os

# 1. INICIALIZAR
spark = SparkSession.builder \
    .appName("Analisis_Nieve_FastTree") \
    .master("local[*]") \
    .config("spark.sql.ansi.enabled", "false") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

base_path = "/home/alumno/reposirotio/Grupo3"
posibles_dirs = glob.glob(f"{base_path}/*2024*")
target_dir = posibles_dirs[0] if posibles_dirs else f"{base_path}/2024.tar"

print(f"Probando Árbol de Decisión (Velocidad) en: {target_dir}")

try:
    # 2. CARGA Y PREPARACIÓN (Incluyendo DayOfYear)
    df = spark.read.option("header", "true").option("inferSchema", "false").csv(target_dir)
    df = df.filter(col("LATITUDE") != "LATITUDE")

    # Ingeniería de Tiempo (Vital para la nieve)
    if "DATE" in df.columns:
        df = df.withColumn("ParsedDate", to_date(col("DATE"), "MM/dd/yyyy"))
        df = df.withColumn("DayOfYear", dayofyear(col("ParsedDate")))
    else:
        df = df.withColumn("DayOfYear", (col("TEMP") * 0) + 1)

    target_col = "SNDP"
    numeric_cols = [
        "DayOfYear", "LATITUDE", "ELEVATION", "TEMP", "DEWP", "SLP", "STP", 
        "VISIB", "WDSP", "MXSPD", "GUST", "MAX", "MIN", "PRCP"
    ]

    select_exprs = []
    for c in numeric_cols + [target_col]:
        clean_col = regexp_replace(trim(col(c)), "[*]", "")
        select_exprs.append(
            when(clean_col == "", 0.0)
            .when(clean_col.cast("double") > 9000, 0.0)
            .when(clean_col.cast("double") == 999.9, 0.0)
            .otherwise(clean_col.cast("double"))
            .alias(c)
        )

    # FRSHTT
    frshtt_col = when(col("FRSHTT").isNull(), "000000").otherwise(col("FRSHTT"))
    frshtt_exprs = [
        substring(frshtt_col, 1, 1).cast("int").alias("Is_Fog"),
        substring(frshtt_col, 2, 1).cast("int").alias("Is_Rain"),
        substring(frshtt_col, 3, 1).cast("int").alias("Is_Snow"),
        substring(frshtt_col, 4, 1).cast("int").alias("Is_Hail"),
        substring(frshtt_col, 5, 1).cast("int").alias("Is_Thunder"),
        substring(frshtt_col, 6, 1).cast("int").alias("Is_Tornado")
    ]

    data = df.select(*select_exprs, *frshtt_exprs).na.fill(0.0)

    # 3. VECTORIZACIÓN
    input_cols = [c for c in data.columns if c != target_col and c != "features"]
    assembler = VectorAssembler(inputCols=input_cols, outputCol="features")
    final_data = assembler.transform(data)

    # 4. SPLIT
    train, test = final_data.randomSplit([0.8, 0.2], seed=42)
    
    # 5. ENTRENAMIENTO: DECISION TREE REGRESSOR
    print("Entrenando Árbol de Decisión Único (Muy rápido)...")
    
    # maxDepth=12:  Hacemos el árbol profundo para que sea inteligente.
    # maxBins=64:   Clave para velocidad y precisión en variables continuas.
    dt = DecisionTreeRegressor(
        featuresCol="features", 
        labelCol=target_col,
        maxDepth=12,
        maxBins=64,
        seed=42
    )
    
    model = dt.fit(train)
    predictions = model.transform(test)

    # 6. RESULTADOS
    r2 = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="r2").evaluate(predictions)
    rmse = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="rmse").evaluate(predictions)

    print("\n" + "="*60)
    print(f"RESULTADOS ÁRBOL DE DECISIÓN ÚNICO")
    print("="*60)
    print(f"R2 (Precisión): {r2:.4f}")
    print(f"RMSE (Error): {rmse:.4f}")
    print("-" * 60)
    
    print("VARIABLES CLAVE (Top 10):")
    # En Decision Tree, featureImportances se calcula igual que en RF
    importances = model.featureImportances
    feat_imp_list = sorted(zip(input_cols, importances), key=lambda x: x[1], reverse=True)
    
    for name, imp in feat_imp_list[:10]:
        print(f"{name:.<25} {imp:.4f} ({imp*100:.1f}%)")
        
    print(f"\nProfundidad real del árbol: {model.depth}")
    print(f"Nodos totales: {model.numNodes}")

except Exception as e:
    print(f"ERROR: {e}")
    import traceback
    traceback.print_exc()

finally:
    if 'spark' in locals() and not spark.sparkContext._jsc.sc().isStopped():
        spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, substring, when, regexp_replace, trim, to_date, dayofyear, cos, sin, radians, lit
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import DecisionTreeRegressor  # <--- SEGUIMOS CON EL VELOCISTA
from pyspark.ml.evaluation import RegressionEvaluator
import glob
import os
import math

# 1. INICIALIZAR (Configuración Robusta)
spark = SparkSession.builder \
    .appName("Analisis_Nieve_FastTree_Max") \
    .master("local[*]") \
    .config("spark.sql.ansi.enabled", "false") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

base_path = "/home/alumno/reposirotio/Grupo3"
posibles_dirs = glob.glob(f"{base_path}/*2024*")
target_dir = posibles_dirs[0] if posibles_dirs else f"{base_path}/2024.tar"

print(f"Optimizando Árbol de Decisión (Física + Profundidad) en: {target_dir}")

try:
    # 2. CARGA
    df = spark.read.option("header", "true").option("inferSchema", "false").csv(target_dir)
    df = df.filter(col("LATITUDE") != "LATITUDE")

    # --- INGENIERÍA DE CARACTERÍSTICAS (EL "DOPAJE" DE DATOS) ---
    
    # 1. Real_Winter_Effect: Soluciona el problema de los hemisferios y el ecuador
    if "DATE" in df.columns:
        df = df.withColumn("ParsedDate", to_date(col("DATE"), "MM/dd/yyyy"))
        df = df.withColumn("DayOfYear", dayofyear(col("ParsedDate")))
        
        # Coseno del día (1 en Ene, -1 en Jul)
        season_cycle = cos((col("DayOfYear") - 1) * (2 * math.pi / 365))
        # Seno de latitud (1 en Polo Norte, -1 en Polo Sur, 0 en Ecuador)
        lat_intensity = sin(radians(col("LATITUDE")))
        
        df = df.withColumn("Real_Winter_Effect", season_cycle * lat_intensity)
    else:
        df = df.withColumn("Real_Winter_Effect", lit(0.0))

    target_col = "SNDP"
    raw_cols = [
        "LATITUDE", "ELEVATION", "TEMP", "DEWP", "SLP", "STP", 
        "VISIB", "WDSP", "MXSPD", "GUST", "MAX", "MIN", "PRCP"
    ]

    select_exprs = []
    for c in raw_cols + [target_col]:
        clean_col = regexp_replace(trim(col(c)), "[*]", "")
        select_exprs.append(
            when(clean_col == "", 0.0)
            .when(clean_col.cast("double") > 9000, 0.0)
            .when(clean_col.cast("double") == 999.9, 0.0)
            .otherwise(clean_col.cast("double"))
            .alias(c)
        )

    # Añadimos la variable maestra de invierno
    select_exprs.append(col("Real_Winter_Effect"))

    # FRSHTT
    frshtt_col = when(col("FRSHTT").isNull(), "000000").otherwise(col("FRSHTT"))
    frshtt_exprs = [
        substring(frshtt_col, 3, 1).cast("int").alias("Is_Snow"), # Nieve
        substring(frshtt_col, 2, 1).cast("int").alias("Is_Rain")  # Lluvia
    ]

    data = df.select(*select_exprs, *frshtt_exprs).na.fill(0.0)

    # 2. Is_Freezing: Pista explícita del punto de congelación (0°C = 32°F)
    data = data.withColumn("Is_Freezing", when(col("MAX") < 32.0, 1.0).otherwise(0.0))

    # 3. VECTORIZACIÓN
    feature_cols = [c for c in data.columns if c != target_col and c != "features"]
    assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
    final_data = assembler.transform(data)

    # 4. SPLIT
    train, test = final_data.randomSplit([0.8, 0.2], seed=42)
    
    # 5. ENTRENAMIENTO: DECISION TREE "SUPERCHARGED"
    print("Entrenando Árbol de Decisión Profundo...")
    
    # MEJORAS AQUÍ:
    # maxDepth=15: Permitimos un árbol gigante para capturar detalles finos.
    # maxBins=128: Permitimos cortes más precisos en la temperatura (ej. cortar en 31.5°F en vez de 30°F).
    dt = DecisionTreeRegressor(
        featuresCol="features", 
        labelCol=target_col,
        maxDepth=15,      # Aumentado de 12 a 15 (Más inteligencia)
        maxBins=128,      # Aumentado de 64 a 128 (Más precisión numérica)
        minInstancesPerNode=5, # Evita que el árbol se vuelva loco con una sola fila (ruido)
        seed=42
    )
    
    model = dt.fit(train)
    predictions = model.transform(test)

    # 6. RESULTADOS
    r2 = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="r2").evaluate(predictions)
    rmse = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="rmse").evaluate(predictions)

    print("\n" + "="*60)
    print(f"RESULTADOS ÁRBOL ÚNICO OPTIMIZADO")
    print("="*60)
    print(f"R2 (Precisión): {r2:.4f}")
    print(f"RMSE (Error): {rmse:.4f}")
    print("-" * 60)
    
    print("VARIABLES CLAVE (Top 12):")
    importances = model.featureImportances
    feat_imp_list = sorted(zip(feature_cols, importances), key=lambda x: x[1], reverse=True)
    
    for name, imp in feat_imp_list[:12]:
        print(f"{name:.<25} {imp:.4f} ({imp*100:.1f}%)")
        
    print(f"\nEstadísticas del árbol:")
    print(f"Profundidad: {model.depth}")
    print(f"Nodos: {model.numNodes}")

except Exception as e:
    print(f"ERROR: {e}")
    import traceback
    traceback.print_exc()

finally:
    if 'spark' in locals() and not spark.sparkContext._jsc.sc().isStopped():
        spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, substring, when, regexp_replace, trim, to_date, dayofyear, cos, lit, signum
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import GBTRegressor
from pyspark.ml.evaluation import RegressionEvaluator
import glob
import os
import math

# 1. INICIALIZAR
spark = SparkSession.builder \
    .appName("Analisis_Nieve_Global_Aware") \
    .master("local[*]") \
    .config("spark.sql.ansi.enabled", "false") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

base_path = "/home/alumno/reposirotio/Grupo3"
posibles_dirs = glob.glob(f"{base_path}/*2024*")
target_dir = posibles_dirs[0] if posibles_dirs else f"{base_path}/2024.tar"

print(f"Entrenando modelo con Corrección Hemisférica en: {target_dir}")

try:
    # 2. CARGA
    df = spark.read.option("header", "true").option("inferSchema", "false").csv(target_dir)
    df = df.filter(col("LATITUDE") != "LATITUDE")

    # --- INGENIERÍA DE CARACTERÍSTICAS CORRECTA (GLOBAL) ---
    
    if "DATE" in df.columns:
        df = df.withColumn("ParsedDate", to_date(col("DATE"), "MM/dd/yyyy"))
        df = df.withColumn("DayOfYear", dayofyear(col("ParsedDate")))
        
        # 1. Estacionalidad Base (Positiva en Ene/Dic, Negativa en Jun/Jul)
        # cos(0) = 1 (1 Enero), cos(pi) = -1 (1 Julio)
        df = df.withColumn("Base_Seasonality", cos((col("DayOfYear") - 1) * (2 * math.pi / 365)))
        
        # 2. Signo de latitud (+1 Norte, -1 Sur, 0 Ecuador)
        # Nota: Casteamos Latitude a double primero para asegurar que signum funcione
        df = df.withColumn("Lat_Sign", signum(col("LATITUDE").cast("double")))
        
        # 3. "Global_Winter": Alinea el invierno de ambos hemisferios hacia el positivo
        df = df.withColumn("Global_Winter", col("Base_Seasonality") * col("Lat_Sign"))
        
    else:
        df = df.withColumn("Global_Winter", lit(0.0))

    target_col = "SNDP"
    raw_cols = [
        "LATITUDE", "ELEVATION", "TEMP", "DEWP", "SLP", "STP", 
        "VISIB", "WDSP", "MXSPD", "GUST", "MAX", "MIN", "PRCP"
    ]

    select_exprs = []
    for c in raw_cols + [target_col]:
        clean_col = regexp_replace(trim(col(c)), "[*]", "")
        select_exprs.append(
            when(clean_col == "", 0.0)
            .when(clean_col.cast("double") > 9000, 0.0)
            .when(clean_col.cast("double") == 999.9, 0.0)
            .otherwise(clean_col.cast("double"))
            .alias(c)
        )

    # Añadimos la nueva variable maestra
    select_exprs.append(col("Global_Winter"))

    # FRSHTT
    frshtt_col = when(col("FRSHTT").isNull(), "000000").otherwise(col("FRSHTT"))
    frshtt_exprs = [
        substring(frshtt_col, 3, 1).cast("int").alias("Is_Snow"),
        substring(frshtt_col, 2, 1).cast("int").alias("Is_Rain")
    ]

    data = df.select(*select_exprs, *frshtt_exprs).na.fill(0.0)
    
    # Is_Freezing sigue siendo útil
    data = data.withColumn("Is_Freezing", when(col("MAX") < 32.0, 1.0).otherwise(0.0))

    # 3. VECTORIZACIÓN
    feature_cols = [c for c in data.columns if c != target_col and c != "features"]
    assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
    final_data = assembler.transform(data)

    # 4. SPLIT
    train, test = final_data.randomSplit([0.8, 0.2], seed=42)
    
    # 5. ENTRENAMIENTO: GBT (Manteniendo la configuración híbrida eficiente)
    print("Entrenando GBT con Lógica Hemisférica...")
    
    gbt = GBTRegressor(
        featuresCol="features", 
        labelCol=target_col,
        maxIter=25,       # Un poco más que antes
        maxDepth=10,      # Profundidad alta
        stepSize=0.15,
        seed=42
    )
    
    model = gbt.fit(train)
    predictions = model.transform(test)

    # 6. RESULTADOS
    r2 = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="r2").evaluate(predictions)
    rmse = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="rmse").evaluate(predictions)

    print("\n" + "="*60)
    print(f"RESULTADOS FINALES (CORRECCIÓN HEMISFERIOS)")
    print("="*60)
    print(f"R2 (Precisión): {r2:.4f}")
    print(f"RMSE (Error): {rmse:.4f}")
    print("-" * 60)
    
    print("IMPORTANCIA DE VARIABLES:")
    importances = model.featureImportances
    feat_imp_list = sorted(zip(feature_cols, importances), key=lambda x: x[1], reverse=True)
    
    for name, imp in feat_imp_list[:12]:
        print(f"{name:.<25} {imp:.4f} ({imp*100:.1f}%)")

except Exception as e:
    print(f"ERROR: {e}")
    import traceback
    traceback.print_exc()

finally:
    if 'spark' in locals() and not spark.sparkContext._jsc.sc().isStopped():
        spark.stop()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, substring, when, regexp_replace, trim, to_date, dayofyear, cos, sin, radians, lit
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.clustering import KMeans
from pyspark.ml.regression import DecisionTreeRegressor
import glob
import os
import math

# ---------------------------------------------------------
# 1. INICIALIZAR SPARK
# ---------------------------------------------------------
spark = SparkSession.builder \
    .appName("Analisis_Nieve_Visualizacion") \
    .master("local[*]") \
    .config("spark.sql.ansi.enabled", "false") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

base_path = "/home/alumno/reposirotio/Grupo3"
posibles_dirs = glob.glob(f"{base_path}/*2024*")
target_dir = posibles_dirs[0] if posibles_dirs else f"{base_path}/2024.tar"

print(f"Recuperando el mejor modelo (K-Means + Tree) para graficar...")

try:
    # ---------------------------------------------------------
    # 2. RECONSTRUIR EL MODELO GANADOR (R2 ~ 0.56)
    # ---------------------------------------------------------
    df = spark.read.option("header", "true").option("inferSchema", "false").csv(target_dir)
    df = df.filter(col("LATITUDE") != "LATITUDE")

    # Conversión geográfica para K-Means
    df = df.withColumn("LATITUDE", col("LATITUDE").cast("double")) \
           .withColumn("LONGITUDE", col("LONGITUDE").cast("double")) \
           .withColumn("ELEVATION", col("ELEVATION").cast("double")) \
           .na.fill(0.0, subset=["LATITUDE", "LONGITUDE", "ELEVATION"])

    # Clustering (Micro-climas)
    geo_assembler = VectorAssembler(inputCols=["LATITUDE", "LONGITUDE", "ELEVATION"], outputCol="geo_features")
    df_geo = geo_assembler.transform(df)
    kmeans = KMeans(featuresCol="geo_features", predictionCol="Climate_Zone", k=50, seed=42)
    km_model = kmeans.fit(df_geo)
    data = km_model.transform(df_geo).drop("geo_features")

    # Ingeniería de Características Física
    if "DATE" in data.columns:
        data = data.withColumn("ParsedDate", to_date(col("DATE"), "MM/dd/yyyy"))
        data = data.withColumn("DayOfYear", dayofyear(col("ParsedDate")))
        # Invierno Real (Latitud + Fecha)
        season_cycle = cos((col("DayOfYear") - 1) * (2 * math.pi / 365))
        lat_intensity = sin(radians(col("LATITUDE")))
        data = data.withColumn("Real_Winter_Effect", season_cycle * lat_intensity)
    else:
        data = data.withColumn("Real_Winter_Effect", lit(0.0))

    target_col = "SNDP"
    raw_cols = ["TEMP", "DEWP", "SLP", "STP", "VISIB", "WDSP", "MXSPD", "GUST", "MAX", "MIN", "PRCP"]

    select_exprs = [col("Climate_Zone"), col("Real_Winter_Effect"), col("LATITUDE"), col("ELEVATION")]
    
    for c in raw_cols + [target_col]:
        clean_col = regexp_replace(trim(col(c)), "[*]", "")
        select_exprs.append(
            when(clean_col == "", 0.0)
            .when(clean_col.cast("double") > 9000, 0.0)
            .when(clean_col.cast("double") == 999.9, 0.0)
            .otherwise(clean_col.cast("double")).alias(c)
        )

    # FRSHTT
    frshtt_col = when(col("FRSHTT").isNull(), "000000").otherwise(col("FRSHTT"))
    frshtt_exprs = [
        substring(frshtt_col, 3, 1).cast("int").alias("Is_Snow"),
        substring(frshtt_col, 2, 1).cast("int").alias("Is_Rain")
    ]

    data = data.select(*select_exprs, *frshtt_exprs).na.fill(0.0)

    # Potencial de Nieve
    data = data.withColumn("Snow_Potential", col("PRCP") * when(col("MAX") < 32, (32 - col("MAX"))).otherwise(0))

    # Vectorizar y Split
    feature_cols = [c for c in data.columns if c != target_col and c != "features"]
    assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
    final_data = assembler.transform(data)
    train, test = final_data.randomSplit([0.8, 0.2], seed=42)
    
    # Entrenar Árbol
    dt = DecisionTreeRegressor(featuresCol="features", labelCol=target_col, maxDepth=15, maxBins=128, minInstancesPerNode=10, seed=42)
    model = dt.fit(train)
    predictions = model.transform(test)

    # ---------------------------------------------------------
    # 3. EXTRACCIÓN DE DATOS PARA GRÁFICOS (AQUÍ FALLABA ANTES)
    # ---------------------------------------------------------
    print("Extrayendo datos para visualización (toPandas)...")
    
    # Muestra para Scatter Plot (Real vs Predicho)
    # Filtramos SNDP > 0 para ver donde hay acción
    sample_df = predictions.select("SNDP", "prediction") \
        .filter("SNDP > 0 OR prediction > 1.0") \
        .sample(False, 0.1, seed=42) \
        .limit(1000) \
        .toPandas()

    # Agregado para Tendencia Temporal (Estacionalidad)
    # Usamos DayOfYear (si existe en el df original)
    if "DayOfYear" in predictions.columns:
        # Recuperamos DayOfYear de features o raw data es complejo, 
        # mejor recalculamos rápido agrupando por Real_Winter_Effect que es proxy de fecha/latitud
        # O simplemente graficamos sample_df ordenado.
        pass
    
    # Haremos un gráfico de serie temporal simple con los primeros 100 puntos de la muestra ordenada
    sample_df = sample_df.sort_values(by="SNDP", ascending=False)

    print(f"Datos extraídos: {len(sample_df)} filas.")

    # ---------------------------------------------------------
    # 4. GENERAR GRÁFICOS
    # ---------------------------------------------------------
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # GRÁFICO 1: REAL vs PREDICCIÓN (Scatter)
    x = sample_df["SNDP"]
    y = sample_df["prediction"]
    
    axes[0].scatter(x, y, alpha=0.5, c='steelblue', edgecolors='k', s=40, label='Muestra Datos')
    
    # Línea ideal
    max_val = max(x.max(), y.max())
    axes[0].plot([0, max_val], [0, max_val], 'r--', lw=2, label='Predicción Perfecta')
    
    axes[0].set_title("Precisión: Realidad vs Modelo", fontsize=14)
    axes[0].set_xlabel("Nieve REAL (pulgadas)", fontsize=12)
    axes[0].set_ylabel("Nieve PREDICHA (pulgadas)", fontsize=12)
    axes[0].legend()
    axes[0].grid(True, alpha=0.3)

    # GRÁFICO 2: ERROR RESIDUAL
    # Muestra cuánto nos equivocamos en cada punto
    residuals = x - y
    axes[1].hist(residuals, bins=30, color='salmon', edgecolor='black', alpha=0.7)
    axes[1].axvline(0, color='k', linestyle='--', linewidth=1)
    
    axes[1].set_title("Distribución de Errores (Residuales)", fontsize=14)
    axes[1].set_xlabel("Error (Real - Predicho)", fontsize=12)
    axes[1].set_ylabel("Frecuencia", fontsize=12)
    axes[1].grid(True, alpha=0.3)
    
    # Nota en el gráfico
    plt.suptitle(f"Análisis Visual del Modelo (R2 ~ 0.56)", fontsize=16)
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"ERROR: {e}")
    import traceback
    traceback.print_exc()

finally:
    if 'spark' in locals() and not spark.sparkContext._jsc.sc().isStopped():
        print("Cerrando Spark...")
        spark.stop()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, substring, when, regexp_replace, trim, to_date, dayofyear, cos, sin, radians, lit
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.clustering import KMeans
from pyspark.ml.regression import DecisionTreeRegressor
import glob
import os
import math

# 1. INICIALIZAR (4GB es suficiente si hacemos downsampling)
spark = SparkSession.builder \
    .appName("Analisis_Nieve_15_Years_Turbo") \
    .master("local[*]") \
    .config("spark.sql.ansi.enabled", "false") \
    .config("spark.driver.memory", "4g") \
    .config("spark.executor.memory", "4g") \
    .getOrCreate()

spark.sparkContext.setLogLevel("ERROR")

base_path = "/home/alumno/reposirotio/Grupo3"

# Leemos TODOS los años
print("1. Cargando histórico completo (15 años)...")
df = spark.read.option("header", "true").option("inferSchema", "false").csv(f"{base_path}/*.tar")
df = df.filter(col("LATITUDE") != "LATITUDE")

# ---------------------------------------------------------
# EL TRUCO DE VELOCIDAD: DOWNSAMPLING
# ---------------------------------------------------------
print("2. Aplicando Downsampling (Eliminando exceso de días sin nieve)...")

# Limpiamos solo la columna SNDP primero para poder filtrar
df = df.withColumn("SNDP_Clean", 
                   when(col("SNDP") == "999.9", 0.0)
                   .otherwise(regexp_replace(col("SNDP"), "[*]", "").cast("double")))

# Separamos los datos
snow_days = df.filter(col("SNDP_Clean") > 0)      # Días interesantes (nos quedamos TODOS)
boring_days = df.filter(col("SNDP_Clean") <= 0)   # Días aburridos (verano/desierto)

# Cogemos solo el 20% de los días aburridos (suficiente para aprender que ahí no hay nieve)
boring_sample = boring_days.sample(False, 0.20, seed=42)

# Unimos de nuevo
df_reduced = snow_days.union(boring_sample)

# Borramos la columna temporal y liberamos memoria
df = df_reduced.drop("SNDP_Clean")

count_original = snow_days.count() + boring_days.count() # Aprox
count_reduced = df.count()

print(f"   Filas originales (estimadas): {count_original}")
print(f"   Filas reducidas para entrenar: {count_reduced}")
print(f"   ¡Hemos reducido el trabajo un {100 - (count_reduced/count_original*100):.1f}%!")

# ---------------------------------------------------------
# FASE NORMAL (K-Means + Ingeniería + Árbol)
# ---------------------------------------------------------

# Conversión Geo
df = df.withColumn("LATITUDE", col("LATITUDE").cast("double")) \
       .withColumn("LONGITUDE", col("LONGITUDE").cast("double")) \
       .withColumn("ELEVATION", col("ELEVATION").cast("double")) \
       .na.fill(0.0, subset=["LATITUDE", "LONGITUDE", "ELEVATION"])

# Clustering Rápido (Sobre el dataset reducido es muy rápido)
print("3. Generando Zonas Climáticas...")
geo_assembler = VectorAssembler(inputCols=["LATITUDE", "LONGITUDE", "ELEVATION"], outputCol="geo_features")
df_geo = geo_assembler.transform(df)
kmeans = KMeans(featuresCol="geo_features", predictionCol="Climate_Zone", k=50, seed=42)
km_model = kmeans.fit(df_geo)
data = km_model.transform(df_geo).drop("geo_features")

# Ingeniería Física
if "DATE" in data.columns:
    data = data.withColumn("ParsedDate", to_date(col("DATE"), "MM/dd/yyyy"))
    data = data.withColumn("DayOfYear", dayofyear(col("ParsedDate")))
    season_cycle = cos((col("DayOfYear") - 1) * (2 * math.pi / 365))
    lat_intensity = sin(radians(col("LATITUDE")))
    data = data.withColumn("Real_Winter_Effect", season_cycle * lat_intensity)
else:
    data = data.withColumn("Real_Winter_Effect", lit(0.0))

# Limpieza Restante
target_col = "SNDP"
# Quitamos columnas que el árbol anterior dijo que no eran importantes (WDSP, SLP, DEWP, GUST)
# Esto acelera AÚN MÁS el proceso
reduced_cols = ["LATITUDE", "ELEVATION", "TEMP", "STP", "VISIB", "PRCP", "MAX", "MIN"]

select_exprs = [col("Climate_Zone"), col("Real_Winter_Effect"), col("LATITUDE"), col("ELEVATION")]

for c in reduced_cols + [target_col]:
    clean_col = regexp_replace(trim(col(c)), "[*]", "")
    select_exprs.append(
        when(clean_col == "", 0.0)
        .when(clean_col.cast("double") > 9000, 0.0)
        .when(clean_col.cast("double") == 999.9, 0.0)
        .otherwise(clean_col.cast("double")).alias(c)
    )

frshtt_col = when(col("FRSHTT").isNull(), "000000").otherwise(col("FRSHTT"))
# Solo nos quedamos con Nieve y Lluvia (las demás no aportaban mucho en el análisis anterior)
frshtt_exprs = [
    substring(frshtt_col, 3, 1).cast("int").alias("Is_Snow"),
    substring(frshtt_col, 2, 1).cast("int").alias("Is_Rain")
]

data = data.select(*select_exprs, *frshtt_exprs).na.fill(0.0)
data = data.withColumn("Is_Freezing", when(col("MAX") < 32.0, 1.0).otherwise(0.0))

# Vectorizar
feature_cols = [c for c in data.columns if c != target_col and c != "features"]
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
final_data = assembler.transform(data)

# Entrenar
print("4. Entrenando Árbol con 15 años de datos optimizados...")
train, test = final_data.randomSplit([0.8, 0.2], seed=42)

dt = DecisionTreeRegressor(
    featuresCol="features", labelCol=target_col, 
    maxDepth=15, maxBins=128, minInstancesPerNode=10, seed=42
)
model = dt.fit(train)
predictions = model.transform(test)

# Resultados
r2 = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="r2").evaluate(predictions)
rmse = RegressionEvaluator(labelCol=target_col, predictionCol="prediction", metricName="rmse").evaluate(predictions)

print("\n" + "="*60)
print(f"RESULTADOS 15 AÑOS (OPTIMIZADO)")
print("="*60)
print(f"R2: {r2:.4f}")
print(f"RMSE: {rmse:.4f}")
print("-" * 60)

spark.stop()
```
